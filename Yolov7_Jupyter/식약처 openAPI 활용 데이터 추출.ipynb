### 식품의약품안전처 open API
- 음식의 영양정보
- https://www.foodsafetykorea.go.kr/api/datasetList.do?cl_cd=API_SRT03&menu_grp=MENU_GRP31&menu_no=17881

```python
import requests as req
import xml.etree.ElementTree as ET
import pandas as pd
```

```python
# 발급받은 api_key값
api_key = ''
```

```python
# 데이터 저장할 리스트 선언
desc_kor_lst = []  # 음식명
group_name_lst = [] # 식품군
maker_name_lst = [] # 제조사명
sampling_region_name_lst = [] # 지역명
serving_size_lst = []  # 중량(g)
nutr_cont1_lst = []  # 열량(kcal)
nutr_cont2_lst = []  # 탄수화물(g)
nutr_cont3_lst = []  # 단백질(g)
nutr_cont4_lst = []  # 지방(g)
nutr_cont5_lst = []  # 당류(g)
nutr_cont6_lst = []  # 나트륨(mg)
nutr_cont7_lst = []  # 콜레스트롤(mg)
nutr_cont8_lst = []  # 포화지방산(g)
nutr_cont9_lst = []  # 트랜스지방(g)
```

```python
# 시작인덱스, 끝인덱스 반복 
# -> 한번에 1000개의 요청만 가능하기때문에 시작인덱스 1000단위로 끊어서 반복시켜줌
for startidx in range(1,90002,1000):
    print(startidx, "시작")
    endidx = startidx+999
    url = f'https://openapi.foodsafetykorea.go.kr/api/{api_key}/I2790/xml/{startidx}/{endidx}'
    response = req.get(url)
    
    # 불러온 XML 데이터에서 음식명, 중량, 칼로리, 탄수화물 등 필요한 태그의 텍스트값 불러오기
    # XML데이터에서 각 row에 순차적 접근
    root = ET.fromstring(response.text)
    for row in root.iter('row'):
        # row에 순차적으로 접근후 필요한태그 접근후 추출
        desc_kor = row.find('DESC_KOR').text # 음식명
        group_name = row.find('GROUP_NAME').text # 식품군
        maker_name = row.find('MAKER_NAME').text # 제조사명
        sampling_region_name = row.find('SAMPLING_REGION_NAME').text # 지역명
        serving_size = row.find('SERVING_SIZE').text # 중량
        nutr_cont1 = row.find('NUTR_CONT1').text # 열량
        nutr_cont2 = row.find('NUTR_CONT2').text # 탄수화물
        nutr_cont3 = row.find('NUTR_CONT3').text # 단백질
        nutr_cont4 = row.find('NUTR_CONT4').text # 지방
        nutr_cont5 = row.find('NUTR_CONT5').text # 당류
        nutr_cont6 = row.find('NUTR_CONT6').text # 나트륨
        nutr_cont7 = row.find('NUTR_CONT7').text # 콜레스트롤
        nutr_cont8 = row.find('NUTR_CONT8').text # 포화지방산
        nutr_cont9 = row.find('NUTR_CONT9').text # 트랜스지방

        # 추출한 데이터 리스트에 저장
        desc_kor_lst.append(desc_kor) # 음식명
        group_name_lst.append(group_name) # 식품군
        maker_name_lst.append(maker_name) # 제조사명
        sampling_region_name_lst.append(sampling_region_name) # 지역명
        serving_size_lst.append(serving_size) # 중량
        nutr_cont1_lst.append(nutr_cont1) # 열량
        nutr_cont2_lst.append(nutr_cont2) # 탄수화물
        nutr_cont3_lst.append(nutr_cont3) # 단백질
        nutr_cont4_lst.append(nutr_cont4) # 지방
        nutr_cont5_lst.append(nutr_cont5) # 당류
        nutr_cont6_lst.append(nutr_cont6) # 나트륨
        nutr_cont7_lst.append(nutr_cont7) # 콜레스트롤
        nutr_cont8_lst.append(nutr_cont8) # 포화지방산
        nutr_cont9_lst.append(nutr_cont9) # 트랜스지방
    
    print(endidx,"끝")
```

```python
# 추출한 정보 데이터 프레임화
food_info_df = pd.DataFrame({'음식명':desc_kor_lst,
                             '식품군':group_name_lst,
                             '제조사명':maker_name_lst,
                             '지역명':sampling_region_name_lst,
                             '중량':serving_size_lst,
                             '열량(kcal)':nutr_cont1_lst,
                             '탄수화물(g)':nutr_cont2_lst,
                             '단백질(g)':nutr_cont3_lst,
                             '지방(g)':nutr_cont4_lst,
                             '당류(g)':nutr_cont5_lst,
                             '나트륨(mg)':nutr_cont6_lst,
                             '콜레스트롤(mg)':nutr_cont7_lst,
                             '포화지방산(g)':nutr_cont8_lst,
                             '트랜스지방(g)':nutr_cont9_lst,
                            })
food_info_df
```

```python
# 엑셀파일로 저장
food_info_df.to_excel('./data/food_info_all.xlsx')
```
