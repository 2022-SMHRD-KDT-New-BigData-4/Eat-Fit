```python
# 음식정보가 담긴 엑셀파일 불러오기
food_df = pd.read_excel('./data/food_info.xlsx')
food_df
```

```python
# 레시피 컬럼 추가 np.nan으로 빈값으로 생성
food_df['레시피'] = np.nan
food_df
```

```python
# 두개의 데이터 프레임에서 식품군, 제조사명이 없고 음식명이 같을때 URL추가하기
for i in tqdm(range(len(food_df))):
    if pd.isna(food_df.iloc[i,2]) and pd.isna(food_df.iloc[i,3]): # 식품군, 제조사명이 없을때
        for j in range(len(url_df)):
            if food_df.iloc[i,1] == url_df.iloc[j,0]: # 음식명이 같을때
                food_df.iloc[i,-1] = url_df.iloc[j,1]
                break
```
